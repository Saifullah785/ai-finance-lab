```python
import pandas_datareader.data as web
import plotly.graph_objects as go
import pandas as pd
from datetime import datetime, timedelta
```

```python
# Define date range (previous 5 years)
end_date = datetime.today()
start_date = end_date - timedelta(days=5*365)
```

# **Gross Domestic Product (GDP):**

 This plot shows the total value of goods and services produced in the US. Look for periods of growth (upward trend) and contraction (downward trend), which indicate economic expansion or recession. You might also see seasonal patterns or impacts of major economic events.

```python
# Fetch data for GDP
try:
    data_gdp = web.DataReader("GDP", 'fred', start_date, end_date)
except Exception as e:
    print(f"Error fetching data for GDP from FRED: {e}")
    data_gdp = pd.DataFrame() # Create an empty DataFrame in case of error

# Plot GDP
if not data_gdp.empty:
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=data_gdp.index, y=data_gdp["GDP"], mode='lines', name="Gross Domestic Product"))
    fig.update_layout(title='Gross Domestic Product (Source: FRED)',
                      xaxis_title='Date',
                      yaxis_title='Gross Domestic Product')
    fig.show()
else:
    print("Could not plot GDP due to data fetching error.")
```

# Task
Analyze and visualize the year-over-year changes of the following economic indicators using Plotly: "GDP", "CPIAUCSL", "UNRATE", "INDPRO", "FEDFUNDS", and "DGS10". The analysis should include calculating the year-over-year percentage change for each indicator and plotting these changes on a single graph for visual comparison.

## Calculate year-over-year change

### Subtask:
Calculate the year-over-year percentage change for each indicator in the `data_combined_resampled` DataFrame.


**Reasoning**:
Calculate the year-over-year percentage change for each column in the resampled combined data and store it in a new DataFrame.



```python
data_YoY = data_combined_resampled.pct_change(periods=12) * 100
display(data_YoY.head())
```

## Prepare data for plotting

### Subtask:
Reshape the data into a format suitable for plotting multiple year-over-year series on the same graph.


**Reasoning**:
Reshape the data into a long format suitable for plotting multiple year-over-year series on the same graph.



```python
data_YoY_melted = data_YoY.melt(var_name='Metric', value_name='Value', ignore_index=False)
display(data_YoY_melted.head())
```

## Visualize year-over-year changes

### Subtask:
Create a Plotly line plot to visualize the year-over-year changes of the selected economic indicators.


**Reasoning**:
Create a Plotly figure and add a scatter trace for each unique metric in the melted data, then update the layout and display the figure.



```python
fig = go.Figure()

for metric in data_YoY_melted['Metric'].unique():
    fig.add_trace(go.Scatter(x=data_YoY_melted.index[data_YoY_melted['Metric'] == metric],
                             y=data_YoY_melted['Value'][data_YoY_melted['Metric'] == metric],
                             mode='lines',
                             name=metric))

fig.update_layout(title='Year-over-Year Change in Economic Indicators (Source: FRED)',
                  xaxis_title='Date',
                  yaxis_title='Year-over-Year Change (%)',
                  legend_title='Indicator')

fig.show()
```

*   A Plotly line plot was generated showing the year-over-year percentage change for "GDP", "CPIAUCSL", "UNRATE", "INDPRO", "FEDFUNDS", and "DGS10" on a single graph, with clear labels and a legend.

## Consider time series analysis

### Subtask:
Depending on the goals, introduce basic time series analysis techniques like decomposition or stationarity testing.


**Reasoning**:
Perform additive decomposition and ADF test on the 'CPIAUCSL' time series and plot the decomposition.



```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt

# Choose the 'CPIAUCSL' indicator
cpi_series = data_fred['CPIAUCSL'].dropna()

# Perform additive decomposition
decomposition = seasonal_decompose(cpi_series, model='additive', period=12)

# Plot the decomposition
fig, axes = plt.subplots(4, 1, sharex=True, figsize=(10, 10))
axes[0].plot(cpi_series)
axes[0].set_title('Original Series')
axes[1].plot(decomposition.trend)
axes[1].set_title('Trend')
axes[2].plot(decomposition.seasonal)
axes[2].set_title('Seasonality')
axes[3].plot(decomposition.resid)
axes[3].set_title('Residuals')
plt.tight_layout()
plt.show()

# Perform Augmented Dickey-Fuller test
adf_test = adfuller(cpi_series)

# Print ADF test results
print('ADF Test Results for CPIAUCSL:')
print(f'Test Statistic: {adf_test[0]}')
print(f'P-value: {adf_test[1]}')
print('Critical Values:')
for key, value in adf_test[4].items():
    print(f'   {key}: {value}')
```

## Summary:

### Data Analysis Key Findings

*   The number of observations for each economic indicator varies significantly due to different data collection frequencies (e.g., GDP has 18 observations while DGS10 has 1246).
*   The correlation matrix shows the linear relationships between the indicators.
*   The heatmap visualization provides a clear visual representation of the correlations, distinguishing between positive and negative relationships.
*   Additive decomposition of the CPIAUCSL series revealed a clear upward trend.
*   The Augmented Dickey-Fuller (ADF) test for CPIAUCSL resulted in a test statistic of -2.6318 and a p-value of 0.0866. With a p-value greater than 0.05, the test suggests that the CPIAUCSL series is likely non-stationary.

### Insights or Next Steps

*   The varying number of observations per indicator should be considered in further analysis, especially when comparing or modeling indicators with different frequencies.
*   Given that the CPIAUCSL series appears non-stationary, differencing or other transformations may be necessary before applying certain time series models that assume stationarity.
